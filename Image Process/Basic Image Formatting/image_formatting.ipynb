###Changing Format

### Contrast and Brightness Enhancement

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/drive/My Drive/Basic Image Formatting/Downloads/4323506-shadow-of-the-erdtree-armors-elden-ring-all-dlc-armors-guide-.jpg')


#print(image.shape[1])           <---- f(i,j)
cv2_imshow(image)

new_image = np.zeros(image.shape, image.dtype)

alpha = 1.5 #<--- Contrast
beta = 1    #<--- Brightness

for a in range(image.shape[0]):                            #<----- Instead of cv.convertScaleAbs(image, alpha=alpha, beta=beta)  |
    for b in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image[a,b,c] = np.clip(alpha*image[a,b,c] + beta, 0, 255)     #<---- g(i,j) = alpha*f(i,j) + beta

cv2_imshow(new_image)
```

```python
import os
import cv2
from google.colab.patches import cv2_imshow

input_destination = '/content/drive/My Drive/Basic Image Formatting/Downloads'
output_destination = '/content/drive/My Drive/Basic Image Formatting/enhanced'

alpha = 1.5 #<--- Contrast
beta = 1    #<--- Brightness

for i in os.listdir(input_destination):

    input_image_name =  os.path.join(input_destination, i)
    input_image = cv2.imread(input_image_name)

    new_image = np.zeros(image.shape, image.dtype)

    if input_image_name.endswith('.jpeg'):
        output_image = os.path.join(output_destination, i.replace('.jpeg', '.png'))

    elif input_image_name.endswith('.jpg'):
        output_image = os.path.join(output_destination, i.replace('.jpg', '.png'))

    new_image = cv2.convertScaleAbs(input_image, alpha=alpha, beta=beta)

    cv2.imwrite(output_image, new_image)
    cv2_imshow(input_image)
    cv2_imshow(new_image)
```

### Histogram Equalization

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/drive/My Drive/Basic Image Formatting/Downloads/4323506-shadow-of-the-erdtree-armors-elden-ring-all-dlc-armors-guide-.jpg')

image =cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

equalised = cv2.equalizeHist(image)

cv2_imshow(image)
cv2_imshow(equalised)
```
