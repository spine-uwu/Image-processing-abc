```python
import os
import cv2 as cv
from google.colab.patches import cv2_imshow



def image_conversion(input_folder,output_folder):
    for filename in os.listdir(input_folder):
      input_path = os.path.join(input_folder, filename)

      if filename.endswith('.jpeg') or filename.endswith('.jpg'):
        output_path = os.path.join(output_folder, filename.replace('.jpeg', '.png').replace('.jpg', '.png'))
        img = cv.imread(input_path)
        cv.imwrite(output_path, img)
        cv2_imshow(img)

input_folder = '/content/drive/My Drive/Histogram/downloaded'
output_folder = "/content/drive/My Drive/Histogram/CLAHEd"
image_conversion(input_folder, output_folder)
```

## CLAHE for single image

```python
import os
import cv2 as cv
from google.colab.patches import cv2_imshow

img = cv.imread('/content/drive/My Drive/Histogram/downloaded/discarded-clothing-from-the-dead-spread-50625939-bcde8a7a5907446ca9246f672639f029.jpg')

gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(gray_img)
equalised = cv.equalizeHist(gray_img)



cv.imwrite('clahe_2.jpg', cl1)
print("raw image")
cv2_imshow(img)
print("Equalized")
cv2_imshow(equalised)
print("CLAHE")
cv2_imshow(cl1)
```

```python
import cv2 as cv
import os
from google.colab.patches import cv2_imshow

input_destination = '/content/drive/My Drive/Histogram/downloaded'
output_destination = "/content/drive/My Drive/Histogram/CLAHEd"


for i in os.listdir(input_destination):
        input_image = os.path.join(input_destination,i)
        gr_img = cv.imread(input_image,0)

        clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        cl1 = clahe.apply(gr_img)

        output_image = os.path.join(output_destination,i)
        cv.imwrite(output_image, cl1)
        cv2_imshow(gr_img)
        cv2_imshow(cl1)

```
